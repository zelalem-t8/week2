```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../data/processed/analyzed_reviews.csv')
plt.figure(figsize=(15, 12))

# 1. Overall Rating Distribution (Pie Chart)
plt.subplot(2, 3, 1)  # Changed to 2x3 grid (6 positions)
rating_counts = df['rating'].value_counts()
plt.pie(rating_counts, 
        labels=rating_counts.index, 
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette("pastel"))
plt.title('Overall Rating Distribution', fontsize=14)

# 2. Overall Sentiment Distribution (Pie Chart)
plt.subplot(2, 3, 2)
sentiment_counts = df['sentiment'].value_counts()
plt.pie(sentiment_counts, 
        labels=sentiment_counts.index,
        autopct='%1.1f%%',
        colors=['#66b3ff','#ff9999','#99ff99'],
        explode=(0.05, 0.05, 0.05))
plt.title('Overall Sentiment Distribution', fontsize=14)

# 3. Bank-wise Sentiment Breakdown
for i, bank in enumerate(df['bank'].unique(), 3):  # Now starts at position 3
    plt.subplot(2, 3, i+1)  # Adjusted to account for 2x3 grid
    bank_data = df[df['bank'] == bank]
    counts = bank_data['sentiment'].value_counts()
    plt.pie(counts,
            labels=counts.index,
            autopct='%1.1f%%',
            colors=['#ff9999','#66b3ff','#99ff99'],
            shadow=True)
    plt.title(f'{bank} Sentiment', fontsize=12)

plt.tight_layout()
plt.show()
```
